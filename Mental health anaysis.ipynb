```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
survey = pd.read_csv("OSMI Mental Health in Tech Survey 2019.csv")
```

```python
survey
```

```python
survey.columns
```

```python
top_10_countries = survey['What country do you *live* in?'].value_counts()[:10].to_frame()
plt.figure(figsize=(10,5))
sns.barplot(top_10_countries['What country do you *live* in?'],top_10_countries.index)
plt.title('Top 10 Countries',fontsize=18,fontweight="bold")
plt.xlabel('')
plt.show()
```

```python
usa = survey.loc[survey['What country do you *live* in?'] == 'United States of America']
top_10_states = usa['What US state or territory do you *live* in?'].value_counts()[:10].to_frame()
plt.figure(figsize=(10,5))
sns.barplot(top_10_states['What US state or territory do you *live* in?'],top_10_states.index)
plt.title('Top 10 US states contributing',fontsize=18,fontweight="bold")
plt.xlabel('')
plt.show()
```

```python
def clean_age(Age):
    if Age>=0 and Age<=100:
        return Age
    else:
        return np.nan
survey['What is your age?'] = survey['What is your age?'].apply(clean_age)
plt.figure(figsize=(25,10))
sns.countplot(survey['What is your age?'].dropna())
plt.title("Age",fontsize=20,fontweight="bold")
plt.show()
```

```python
diagnosis = survey['Have you ever been *diagnosed* with a mental health disorder?'].value_counts()[:10].to_frame()
plt.figure(figsize=(10,5))
sns.barplot(diagnosis['Have you ever been *diagnosed* with a mental health disorder?'],diagnosis.index)
plt.title('Have you ever been diagnosed with a mental health disorder?',fontsize=18,fontweight="bold")
plt.xlabel('')
plt.show()
```

```python
response= survey['Describe the circumstances of the badly handled or unsupportive response.'].dropna()
response
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from wordcloud import WordCloud, STOPWORDS
motive=response.str.lower().str.replace(r'\|', ' ').str.cat(sep=' ')
words=nltk.tokenize.word_tokenize(motive)
word_dist = nltk.FreqDist(words)
stopwords = nltk.corpus.stopwords.words('english')
words_except_stop_dist = nltk.FreqDist(w for w in words if w not in stopwords) 
wordcloud = WordCloud(stopwords=STOPWORDS,background_color='black').generate(" ".join(words_except_stop_dist))
plt.imshow(wordcloud)
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.axis('off')
plt.show()
```
